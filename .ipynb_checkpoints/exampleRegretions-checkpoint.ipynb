```python
import torch
import numpy as np
import torch.nn as nn  ## neural network
import torch.optim as optim
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

<div class='alert alert-info'>
<h1>Regresión Lineal</h1>
</div>

```python
car_prices = [5, 6, 7, 8, 9, 10]
units_sold = [8.5, 8, 7.5, 7, 6.5, 6]

plt.scatter(car_prices, units_sold)
plt.xlabel('precios')
plt.ylabel('unidades')
plt.show()
```

```python
prices_array = np.array(car_prices).reshape(-1, 1)
units_array = np.array(units_sold).reshape(-1, 1)
```

```python
prices = torch.from_numpy(prices_array).float().requires_grad_(True)
units = torch.from_numpy(units_array).float()
```

```python
# Creamos el modelo
model = nn.Linear(1, 1)
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.015)

losses = []
iterations = 2000

for i in range(iterations):
    # Forward pass
    pred = model(prices)
    loss = loss_function(pred, units)
    losses.append(loss)
    
    # Backward pass
    optimizer.zero_grad() # reinicio gradientes
    loss.backward()
    optimizer.step()
    
print(loss)

plt.plot(range(iterations), losses)
plt.xlabel('Interaction')
plt.ylabel('Loss')
plt.show()
```

```python
x = torch.Tensor([[7]])
p = model(x)
p
```

<div class='alert alert-info'>
<h1>Regresión Logística</h1>
</div> 

```python
n = 100
h = n//2
dim = 2

data = np.random.randn(n, dim)*3
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
data[:h, :] = data[:h, :] - 3*np.ones((h, dim)) # primer mitad centrada en -3, -3
data[h:, :] = data[h:, :] + 3*np.ones((h, dim)) # primer mitad centrada en 3, 3 
```

```python
colors = ['blue', 'orange']
color = np.array([colors[0]]*h + [colors[1]]*h).reshape(n)
plt.scatter(data[:,0], data[:,1], c=color, s=75, alpha=0.7)
plt.show()
```

```python
target = np.array([0]*h + [1]*h).reshape(n, 1)

x = torch.from_numpy(data).float().requires_grad_(True)
y = torch.from_numpy(target).float()
```

```python
model = nn.Sequential(
            nn.Linear(2,1),
            nn.Sigmoid()
        )
```

```python
loss_function = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
iterations = 1000
losses = []

for i in range(iterations):
    # Forward pass
    result = model(x)
    loss = loss_function(result, y)
    losses.append(loss.data)
    
    # Backward pass
    optimizer.zero_grad() # reinicio gradientes
    loss.backward()
    optimizer.step()
    
print(loss)

plt.plot(range(iterations), losses)
plt.xlabel('Interaction')
plt.ylabel('Loss')
plt.show()
```

```python
blue = torch.Tensor([[-5, -6]])
prediction = model(blue)
prediction
```

Como es menor a 0.5 entonces predijo correctamente que es azul.

```python
w = list(model.parameters())
w0 = w[0].data.numpy() ## [pendiente, nose]
w1 = w[1].data.numpy() ## b

plt.scatter(data[:,0], data[:,1], c=color, s=75, alpha=0.6)

x_axis = np.linspace(-10, 10, n)
y_axis = -(w1[0] + x_axis*w0[0][0]) / w0[0][1]
plt.plot(x_axis, y_axis, 'g--')
plt.show()
```
